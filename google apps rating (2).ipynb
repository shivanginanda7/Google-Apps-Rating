### Filling null values with aggregate functions

```python
def impute_median(series):
    return series.fillna(series.median())
```

```python
ds1.Rating=ds1['Rating'].transform(impute_median)
```

```python
ds1.isnull().sum()
```

```python
#Mode of categorical values
print(ds1['Type'].mode())
print(ds1['Current Ver'].mode())
print(ds1['Android Ver'].mode())
```

```python
#Fill the missing categorical values with mode
ds1['Type'].fillna(str(ds1['Type'].mode().values[0]),inplace=True)
ds1['Current Ver'].fillna(str(ds1['Current Ver'].mode().values[0]),inplace=True)
ds1['Android Ver'].fillna(str(ds1['Android Ver'].mode().values[0]),inplace=True)
```

```python
ds1.isnull().sum()
```

```python
ds1['Price']=ds1['Price'].apply(lambda x: str(x).replace('$','')if'$' in str(x)else str(x))
ds1['Price']=ds1['Price'].apply(lambda x: float(x))
ds1['Reviews']=pd.to_numeric(ds1['Reviews'], errors='coerce')
```

```python
ds1['Installs']=ds1['Installs'].apply(lambda x: str(x).replace('+','')if'+'in str(x)else str(x))
ds1['Installs']=ds1['Installs'].apply(lambda x: str(x).replace(',','')if','in str(x)else str(x))
ds1['Installs']=ds1['Installs'].apply(lambda x: float(x))
```

## Data Visualization

```python
grp= ds1.groupby('Category')
x=grp['Rating'].agg(np.mean)
print(x)
z=grp['Reviews'].agg(np.sum)
print(z)
y=grp['Price'].agg(np.mean)
print(y)
```

```python
plt.plot(x)
```

```python
plt.figure(figsize=(12,5))
plt.plot(x,'ro')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(x,'ro',color='b')
plt.xticks(rotation=90)
plt.title('Category wise Rating')
plt.xlabel('Category')
plt.ylabel('Rating')
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(z,'r--',color='b')
plt.xticks(rotation=90)
plt.title('Category wise Pricing')
plt.xlabel('Category')
plt.ylabel('Prices')
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(z,'g^',color='y')
plt.xticks(rotation=90)
plt.title('Category wise Reviews')
plt.xlabel('Category')
plt.ylabel('Reviews')
plt.show()
```
